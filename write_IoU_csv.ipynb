```python
# Output as csv file
import os
import csv
path = '/home/qeek/yolo/IoU/'
file_list = os.listdir(path)
res = []

for file in file_list:
    if file.startswith('.'):
        file_list.remove(file)

with open('/home/qeek/yolo/CSV_file/IoU.csv', 'w') as output:
    writer = csv.writer(output)
    writer.writerow(['IoU'])
    
for i in range(1,len(file_list)+1):
    res = []       
    s = ''
 
    if i < 10:
        file = '/home/qeek/yolo/IoU/validation_00{}.txt'.format(i)
    elif i >= 10 and i < 100:
        file = '/home/qeek/yolo/IoU/validation_0{}.txt'.format(i) 
    elif i >= 100:
        file = '/home/qeek/yolo/IoU/validation_{}.txt'.format(i)
   
    with open(file,'r') as f:
        content = f.read()
        lines = content.split('\n')
        n = len(lines)-1
        for j in range(n):
            name = lines[j].split(' ')[0]   
            iou = lines[j].split(' ')[1]
            s += name + ' '     
            s = s + ',' +  iou + ' ; '
        res.append(s[:-3])


    with open('/home/qeek/yolo/CSV_file/IoU.csv', 'a') as output:
        writer = csv.writer(output)
        writer.writerow(res)
```

```python
# Combine with other csv file
import pandas as pd

df1 = pd.read_csv('/home/qeek/yolo/CSV_file/yolo_result.csv')
df2 = pd.read_csv('/home/qeek/yolo/CSV_file/IoU.csv')

result = pd.concat([df1, df2], axis=1, join_axes=[df1.index])
```
