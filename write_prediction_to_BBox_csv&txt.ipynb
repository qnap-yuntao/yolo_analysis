```python
# Output as csv file
import os
import csv

path = '/home/qeek/yolo/Prediction_BBOX/Predicting_BBOX_val'
mydict = {'QNAP Logo ':'0','Big Heart ':'3', 'Finger Heart ':'1', 'Hands Heart ':'2'}

file_list = os.listdir(path)
for file in file_list:
    if file.startswith('.'):
        file_list.remove(file)

with open('/home/qeek/yolo/CSV_file/Predicting_BBOX.csv', 'a') as output:
    writer = csv.writer(output)
    writer.writerow(['img_name','predict_label'])
#     writer.writerow(['img_name','predict_label','ground_truth_label','IoU', 'Iou_avg','precision','recall','f_score','accuracy'])
        
for i in range(1,len(file_list)+1):
    res = []
    if i < 10:
        file = 'validation_00{}.txt'.format(i)
    elif i >= 10 and i < 100:
        file = 'validation_0{}.txt'.format(i) 
    elif i >= 100:
        file = 'validation_{}.txt'.format(i)
    res.append(file[:-4])
    s = ''
#     print(file)
    with open(os.path.join(path, file), 'r') as f:
        content = f.read()
        lines = content.split('\n')
        n = len(lines)-1
        name = ''
        for i in range(n) :
            name = lines[i].split(',')
            name = name[0].split('0.')[0]
            if name in mydict:
                s = s + (mydict[name]) + ' '
                
            box = lines[i].split('(')[1]
            box = box[:-1]
            box = box.split(', ')
            for j in range(len(box)):
                box[j] = int(box[j].split('.')[0])
            [x,y,w,h] = box
            xmin = x-int(w/2.0)
            ymin = y-int(h/2.0)
            xmax = x+int(w/2.0)
            ymax = y+int(h/2.0)   
            s = s + ',' +  str(xmin) + ' ' + str(ymin) + ' ' + str(xmax) + ' ' + str(ymax) + ' ; '
        res.append(s[:-3])
#         print(res)
        with open('/home/qeek/yolo/CSV_file/Predicting_BBOX.csv', 'a') as output:
            writer = csv.writer(output)
            writer.writerow(res)
```
